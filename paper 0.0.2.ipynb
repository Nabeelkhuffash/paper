```python
# class based cclassifier, still missing some features but have the basic functionality
class Classifier:
    def __init__ (self, data_frame, class_name="class", encoding="count", classifier_chain = None):
        self.data_frame = data_frame
        self.class_name = class_name

        self.x = None
        self.y = None
        
        self.training_data_x = None
        self.training_data_y = None

        self.validation_data_x = None
        self.validation_data_y = None

        self.test_data_x = None
        self.test_data_y = None

        if classifier_chain != None:
            self.class_name = classifier_chain[0]
        self.classifier_chain = classifier_chain

        self.childs = {}

        self.classifier = None

        self.chain_training_data_y = {}
        self.chain_validation_data_y = {}
        self.chain_test_data_y = {}
    def get_data_frame(self):
        return self.data_frame
    
    @staticmethod
    def build_kmers(sequence, k_size):
        kmers = []
        #k_size is the size of each word
        n_kmers = len(sequence) - k_size + 1

        for i in range(n_kmers):
            kmer = sequence[i:i + k_size]
            kmers.append(kmer)

        return kmers

    @staticmethod
    def encodeWithKemersAndBagOfWords (data_frame, dna_name, k_size, encoding=None):
        #encode the DNA and return it and sparse matrix


        #this function assume that the data frame has 2 columns "DNA" which contain the DNA string
        #and "Class" which contain the class we want to train and predict on

        #first step is to build k-mers for each dna string which can be done by the function we already created
        for i in range (len(data_frame)):
            data_frame[dna_name].iloc[i]=Classifier.build_kmers(data_frame[dna_name].iloc[i], k_size)

        #join the words together so they create a paragraph (a single string)
        words = list(data_frame[dna_name])
        for i in range (len(words)):
            words[i]= " ".join(words[i])

        #after having the DNA ready as paragraph we can use natural language encoders to encode it
        if encoding=="tfidf" or encoding == 2:
            from sklearn.feature_extraction.text import TfidfVectorizer
            cv = TfidfVectorizer()
            x = cv.fit_transform(words)
        elif encoding =="hashing" or encoding ==3:
            n_features = 16000
            from sklearn.feature_extraction.text import HashingVectorizer
            vectorizer = HashingVectorizer(n_features=n_features)
            x = vectorizer.transform(words)
        else:
            from sklearn.feature_extraction.text import CountVectorizer
            cv = CountVectorizer()
            x = cv.fit_transform(words)
        return x
        #x is the sparse matrix after encoding and y is class list
        
    @staticmethod
    def delete_row_csr(mat, i):
        # a function to delete a a row from sparse matrix, it used in crossValidation function
        from scipy.sparse import csr_matrix
        n = mat.indptr[i+1] - mat.indptr[i]
        if n > 0:
            mat.data[mat.indptr[i]:-n] = mat.data[mat.indptr[i+1]:]
            mat.data = mat.data[:-n]
            mat.indices[mat.indptr[i]:-n] = mat.indices[mat.indptr[i+1]:]
            mat.indices = mat.indices[:-n]
        mat.indptr[i:-1] = mat.indptr[i+1:]
        mat.indptr[i:] -= n
        mat.indptr = mat.indptr[:-1]
        mat._shape = (mat._shape[0]-1, mat._shape[1])
    
    def crossValidationSets(self, x, y, test_size = 0.11111, seed = 123123, splits=1):
        y = y.to_numpy()
        y = y.tolist()
        #we are using shuffleSplit which is the most randomized model and have no constrains on splits number 
        from sklearn.model_selection import ShuffleSplit
        import numpy as np
        from scipy.sparse import csr_matrix
        ss = ShuffleSplit(n_splits=splits, test_size=test_size, random_state=seed)

        #shuffleSplit don't actually split data, but it return a list of indexes that can be used to split the data
        #split the data based on it

        trainMatrix =[]
        validateMatrix =[]
        trainMatrixY =[]
        validateMatrixY =[]
        #for each split we have we will make a new list
        for train_index, validate_index in ss.split(x):

            #you can't build sparse matrix by adding new rows to it, so I coppied the original matrix and removed the unwanted
            #rows from it
            trainSetY =y.copy()
            trainSet =x.copy()
            for i in range (len(validate_index)):
                self.delete_row_csr(trainSet, validate_index[i])
                trainSetY.pop(validate_index[i])
                for j in range (len(validate_index)):
                    if validate_index[j]>validate_index[i]:
                        validate_index[j] = validate_index[j]-1
            trainMatrix.append(trainSet)
            trainMatrixY.append(trainSetY)


            validateSetY =y.copy()
            validateSet =x.copy()
            for i in range (len(train_index)):
                self.delete_row_csr(validateSet, train_index[i])
                validateSetY.pop(train_index[i])
                for j in range (len(train_index)):
                    if train_index[j]>train_index[i]:
                        train_index[j] = train_index[j]-1
            validateMatrix.append(validateSet)
            validateMatrixY.append(validateSetY)
        
        self.training_matrix_x = trainMatrix
        self.training_matrix_y = trainMatrixY
        self.validation_matrix_x = validateMatrix
        self.validation_matrix_y = validateMatrixY
        
        return trainMatrix, validateMatrix, trainMatrixY, validateMatrixY

    def split_data(self, seed = 42):
        from sklearn.model_selection import train_test_split
        if self.classifier_chain != None:
            for value in self.classifier_chain:
                training_data_y, test_data_y = train_test_split(self.data_frame[value], test_size = 0.1, random_state=42)
                training_data_y, validation_data_y = train_test_split(training_data_y, test_size = 0.111111, random_state=seed)
                self.chain_training_data_y[value] = training_data_y.values
                self.chain_validation_data_y[value] = validation_data_y.values
                self.chain_test_data_y[value] = test_data_y.values
            
        self.training_data_x, self.test_data_x,  self.training_data_y, self.test_data_y = train_test_split(self.x, self.y, test_size = 0.1, random_state=42)
        self.training_data_x, self.validation_data_x,  self.training_data_y, self.validation_data_y = train_test_split(self.training_data_x, self.training_data_y, test_size = 0.111111, random_state=seed)
    
    def buildClassifier(self, seed = 42, classifier_type = 1, kernel="ploy",degree=1, C = 1, coef0=1, gamma=1, n_estimators=30, max_depth=25, max_leaf_nodes = 100, alpha=1):
        if self.training_data_x  == None or self.training_data_y == None:
            self.split_data(seed = seed)
        try:
            if  classifier_type ==2:
                from sklearn.ensemble import RandomForestClassifier
                classifier = RandomForestClassifier(n_estimators = n_estimators,max_depth=max_depth,max_leaf_nodes=max_leaf_nodes, random_state =42)
                classifier.fit(self.training_data_x, self.training_data_y)
    
            elif classifier_type ==3:
                from sklearn import svm
                classifier = svm.SVC(kernel=kernel, degree=degree, coef0 =coef0, C=C, gamma = gamma)
                classifier.fit(self.training_data_x, self.training_data_y)
                
            else:
                from sklearn.naive_bayes import MultinomialNB
                classifier = MultinomialNB(alpha=alpha)
                classifier.fit(self.training_data_x, self.training_data_y)
        except ValueError as ve: #catch cases where 1 class is passed
                return FakeClassifier(self.training_data_y[0])
        return classifier
    
    def findAccuracyClassifier (self, train = False, validate= False, test= False, seed = 42, classifier_type =1, C =1, coef0=1, gamma = 1, kernel="poly", degree = 1, n_estimators= 30, max_depth=25,max_leaf_nodes=100, alpha = 1):
        classifier = self.classifier
        if classifier == None:
            classifier = self.buildClassifier(seed=seed, classifier_type=classifier_type, degree = degree, gamma=gamma, kernel = kernel, C = C, coef0=coef0, n_estimators=n_estimators, max_depth=max_depth, max_leaf_nodes=max_leaf_nodes, alpha=alpha)
            self.classifier = classifier
        
        y = None
        predicted_y = None
        if self.classifier_chain != None:
            predicted_y = []
            for item in self.validation_data_x:
                predicted_y.append(self.predict_value_chain(value = item, seed=seed, classifier_type=classifier_type, degree = degree, gamma=gamma, kernel = kernel, C = C, coef0=coef0, n_estimators=n_estimators, max_depth=max_depth, max_leaf_nodes=max_leaf_nodes, alpha=alpha))
            y = self.chain_validation_data_y[self.classifier_chain[-1]]
        elif validate == True:
            #use the classifier to predict the y on validate set and compare it with original y using sklearn matrics
            predicted_y = classifier.predict(self.validation_data_x)
            predicted_y = predicted_y.tolist()
            y = self.validation_data_y

        elif test == True and validate == False:
            #use the classifier to predict the y on train set and compare it with original y using sklearn matrics
            predicted_y = classifier.predict(self.test_data_x)
            predicted_y = predicted_y.tolist()
            y = self.test_data_y
        else:
            return "invalid parameters!"
        
        from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, confusion_matrix
        return accuracy_score(y, predicted_y), f1_score(y, predicted_y, average='macro'), precision_score(y, predicted_y, average='macro'), recall_score(y, predicted_y, average='macro'), confusion_matrix(y, predicted_y)

    def predict_value_chain(self, value, seed = 42, classifier_type = 1, kernel="ploy",degree=1, C = 1, coef0=1, gamma=1, n_estimators=30, max_depth=25, max_leaf_nodes = 100, alpha=1):
        #import pdb
        #pdb.set_trace()
        classifier = self.classifier
        if classifier == None:
            classifier = self.buildClassifier(seed=seed, classifier_type=classifier_type, degree = degree, gamma=gamma, kernel = kernel, C = C, coef0=coef0, n_estimators=n_estimators, max_depth=max_depth, max_leaf_nodes=max_leaf_nodes, alpha=alpha)
        if self.classifier_chain == None or len(self.classifier_chain) < 2:
            return classifier.predict(value)
        result = classifier.predict(value)[0]
        if result in self.childs:
            return self.childs[result].predict_value_chain(value, seed=seed, classifier_type=classifier_type, degree = degree, gamma=gamma, kernel = kernel, C = C, coef0=coef0, n_estimators=n_estimators, max_depth=max_depth, max_leaf_nodes=max_leaf_nodes, alpha=alpha)
        new_classifier = Classifier(self.data_frame[self.data_frame[self.class_name] == result], classifier_chain = self.classifier_chain[1:])
        new_x = []
        new_y = []
        y_set = self.chain_training_data_y[self.classifier_chain[1]]
        for key, v in self.chain_training_data_y.items():
            if key != new_classifier.class_name:
                new_classifier.chain_training_data_y[key]=[]
        for i in range(len(self.training_data_y)):
            if self.training_data_y[i] == result:
                new_x.append(self.training_data_x[i])
                new_y.append(y_set[i])
                for key, v in new_classifier.chain_training_data_y.items():
                    new_classifier.chain_training_data_y[key].append(self.chain_training_data_y[key][i])
        from scipy.sparse import vstack
        new_classifier.training_data_x = vstack(new_x)
        new_classifier.training_data_y = new_y
        self.childs[result] = new_classifier
        return new_classifier.predict_value_chain(value, seed=seed, classifier_type=classifier_type, degree = degree, gamma=gamma, kernel = kernel, C = C, coef0=coef0, n_estimators=n_estimators, max_depth=max_depth, max_leaf_nodes=max_leaf_nodes, alpha=alpha)
```

```python
class FakeClassifier:
    def __init__ (self, prediction_result):
        self.prediction_result = prediction_result
    def predict(self, x):
        return [self.prediction_result]
```

```python
def get_average_from_cross_validation(cross_validation_result): #temp function, should be included in the class
    average_of_cross_validation = []
    for i in range(0, 4):
        my_temp = 0;
        for j in range (0, len(cross_validation_result)):
            my_temp+=cross_validation_result[j][i];
        average_of_cross_validation.append(my_temp/len(cross_validation_result))
    return average_of_cross_validation
```
